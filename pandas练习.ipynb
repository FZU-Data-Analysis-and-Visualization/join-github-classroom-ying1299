```python
#2.
import pandas as pd
a= {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1= pd.DataFrame(a)
print("原始数据：")
print(df1)
#1.
print("1. 选取 score 和 gender 两列：")
aa = df1[['score', 'gender']]
print(aa)

#2. 
print("2.选取索引 0、2 的行：")
bb= df1.loc[[0, 2]]
print(bb)

#3. 
print("3.筛选 score>80 的行：")
cc= df1[df1['score'] > 80]
print(cc)

#4.
print("4. 新增grade 列：")
def get_grade(x):
    if x>= 90:
        return '优秀'
    else:
        return '良好'
df1['grade'] = df1['score'].apply(get_grade)
print(df1)

# 5. 
print("5. 删除 gender 列：")
df1= df1.drop(columns=['gender'])
print(df1)
```

```python
#3.
import pandas as pd
import numpy as np

#1:创建含缺失值的DataFrame
aa= {"x": [1, None, 3], "y": [None, 5, 6]}
df = pd.DataFrame(aa)
print(df)

#检测缺失值（返回布尔值矩阵，True表示缺失）
print("\n检测缺失值：")
print(df.isnull())

#统计每列缺失值数量
print("\n每列缺失值数量：")
print(df.isnull().sum())

# 用0填充缺失值
bb = df.fillna(0)
print("用0填充缺失值后的结果：")
print(bb)

# 删除含缺失值的行
cc= df.dropna()
print("\n删除含缺失值的行后的结果：")
print(cc)
```

```python
#4.
import pandas as pd
aa= {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(aa)

print("1：按 gender 分组，求 score 的均值和最大值")
bb = df1.groupby('gender')['score'].agg(['mean', 'max'])
print(bb)

print("2：按gender分组，统计每组人数")
cc = df1.groupby('gender').size()
print(cc)
```

```python
import pandas as pd
aa= {
    'name': ['Tom', 'Jerry', 'Mike'],
    'score': [85, 92, 78],
    'gender': ['男', '女', '男']
}
df1 = pd.DataFrame(aa)

print("1：按score升序排序")
df_sorted = df1.sort_values(by='score', ascending=True)
print(df_sorted)

print("2：添加重复数据并去重")
# 1. 添加一行重复数据（Tom, 85, 男）
new_row = pd.DataFrame([{'name': 'Tom', 'score': 85, 'gender': '男'}])
df_with_dup = pd.concat([df1, new_row], ignore_index=True)
print("添加重复数据后：")
print(df_with_dup)

# 2. 去重（删除完全相同的重复行）
df_dedup = df_with_dup.drop_duplicates()
print("\n去重后：")
print(df_dedup)
```
